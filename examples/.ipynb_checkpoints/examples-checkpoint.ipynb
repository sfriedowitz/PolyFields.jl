```julia
import Pkg
Pkg.activate("../")

using PolyFields
using Plots
```

# Diblock

## Lamellar

```julia
dims = (128, 1, 1)
cell = UnitCell(1, :lamellar, 5.0)

amon = Monomer(id = 1, vol = 1.0)
bmon = Monomer(id = 2, vol = 1.0)
sys = FieldSystem(dims, cell; monomers = [amon, bmon], ensemble = Canonical)

chain = Diblock(amon, bmon, 100, 0.5, 1.0, 1.0, 100)
add_species!(sys, chain, 1.0)

itx = FloryInteraction()
set_interaction!(itx, amon, bmon, 0.25)
add_interaction!(sys, itx)

fieldinit!(sys, "seeds/lam1d.json")
```

```julia
opts = SCFTOptions(nsteps = 2500, nsout = 100, ftol = 1e-5, stol = 5e-5)

updater = MomentumUpdater(method = :MOM, lam = 0.025)
domain = DomainUpdater(nper = 10, nskip = 100, lam = 0.25, rlxn = 0.9)

scft!(sys, updater; domain = domain, opts = opts)
```

```julia
plot(sys.density[1][:,1,1])
```

## Cylinder

```julia
dims = (128, 128, 1)
cell = UnitCell(2, :hexagonal, 5.0)

amon = Monomer(id = 1, vol = 1.0)
bmon = Monomer(id = 2, vol = 1.0)
sys = FieldSystem(dims, cell; monomers = [amon, bmon], ensemble = Canonical)

chain = Diblock(amon, bmon, 100, 0.3, 1.0, 1.0, 100)
add_species!(sys, chain, 1.0)

itx = FloryInteraction()
set_interaction!(itx, amon, bmon, 0.3)
add_interaction!(sys, itx)

fieldinit!(sys, "seeds/cyl2d.json")
```

```julia
opts = SCFTOptions(nsteps = 10000, nsout = 1000, ftol = 1e-5, stol = 5e-5)

updater = MomentumUpdater(method = :MOM, lam = 0.1)
domain = DomainUpdater(nper = 10, nskip = 100, lam = 0.25, rlxn = 0.9)

scft!(sys, updater; domain = domain, opts = opts)
```

```julia
heatmap(sys.density[1][:,:,1])
```

# Blend

```julia
dims = (256, 1, 1)
cell = UnitCell(1, :lamellar, 20.0)

amon = Monomer(id = 1, vol = 2.0)
bmon = Monomer(id = 2, vol = 2.0)
cmon = Monomer(id = 3, vol = 1.0)

sys = FieldSystem(dims, cell; monomers = [amon, bmon, cmon], ensemble = Canonical)

point = Point(cmon)
achain = Homopolymer(amon, 100, 1.0)
bchain = Homopolymer(bmon, 100, 1.0)

add_species!(sys, point, 0.4)
add_species!(sys, achain, 0.3)
add_species!(sys, bchain, 0.3)

itx = FloryInteraction()
set_interaction!(itx, amon, bmon, 0.5)
set_interaction!(itx, amon, cmon, 1.25)
set_interaction!(itx, bmon, cmon, 1.25)
add_interaction!(sys, itx)

fieldinit!(sys; scale = 0.05, seed = 12345)
```

```julia
opts = SCFTOptions(nsteps = 5000, nsout = 500, ftol = 1e-5, stol = 5e-5)

updater = MomentumUpdater(method = :MOM, lam = 0.01)

scft!(sys, updater; opts = opts)
```

```julia
plot(sys.density[1][:,1,1])
plot!(sys.density[2][:,1,1])
plot!(sys.density[3][:,1,1])
```
